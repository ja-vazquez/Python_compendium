# Bisection

```python
# Python program for implementation 
# of Bisection Method for 
# solving equations 

#The function is x^2  - 2 
#The value of root is : sqrt(2)
def func(x): 
    return x**2 - 2
```

```python
# Prints root of func(x) 
# with error of EPSILON 
def bisection(a,b, tol=10**-6, N=50): 
  
    if (func(a) * func(b) >= 0): 
        print("You have not assumed right a and b\n") 
        return
    i=0
    c_arr = []
    f_arr = []
    c = a 
    while ((b-a) >= tol): 
        # Find middle point 
        c = (a+b)/2
        
        c_arr.append(c)
        f_arr.append(func(c))
        print('%d, %.5f, %.5f, \t , %.5f, %.5f '%(i, a, b, c, func(c)) )
        
        # Check if middle point is root 
        if abs(func(c)) <= tol:  #(func(c) ==0): 
            break
        
        # Decide the side to repeat the steps 
        if (func(c)*func(a) < 0):  #use sign function
            b = c 
        else: 
            a = c 
        i+=1
        if i==N: break
    print("\nThe value of root is : ","%.5f"%c) 
    return c_arr, f_arr
```

```python
a = 0
b = 2
c, f = bisection(a, b) 
```

```python
x = np.linspace(0, 2, 50)
plt.plot(x, func(x))
plt.hlines(0, 0, 2)
plt.plot(c, f, 'ro')
plt.plot(a, func(a), 'bo', markersize=12)
plt.plot(b, func(b), 'bo', markersize=12)
plt.grid()
for i, (j,k) in enumerate(zip(c,f)):
    plt.text(j, k-0.3, '%d'%(i+1))
```

# False Position

```python
# Python3 implementation of False Position 
  
def FalsePos( a , b, tol=10**-5, N=50): 
    if func(a) * func(b) >= 0: 
        print("You have not assumed right a and b") 
        return -1
      
    j=0
    c = a # Initialize result 
    c_arr = []
    f_arr = []
    while ((b-a) >= tol): 
    #for i in range(N): 
        # Find the point that touches x axis 
        c = (a * func(b) - b * func(a))/ (func(b) - func(a)) 
          
        # Check if the above found point is root 
        if abs(func(c)) <= tol: #(func(c) ==0): 
            break
        
        c_arr.append(c)
        f_arr.append(func(c))
        print('%d, %.5f, %.5f, \t , %.5f, %.5f '%(j, a, b, c, func(c)) )
        
        # Decide the side to repeat the steps 
        if func(c) * func(a) < 0: 
            b = c 
        else: 
            a = c 
        j+=1
        if j==N: break
    print("The value of root is : " , '%.5f' %c) 
    return c_arr, f_arr
```

```python
# Fixed Point Iteration Method
# Importing math to use sqrt function
import math

def f(x):
    return x**3 + x**2 -1

# Re-writing f(x)=0 to x = g(x)
def g(x):
    return 1/math.sqrt(1+x)
```

```python
# Implementing Fixed Point Iteration Method
def fixedPointIteration(x0, e, N):
    print('\n\n*** FIXED POINT ITERATION ***')
    step = 1
    flag = 1
    condition = True
    c1_arr = []
    f1_arr = []
    while condition:
        x1 = g(x0)
        print('Iteration-%d, x1 = %0.6f and f(x1) = %0.6f' % (step, x1, f(x1)))
        c1_arr.append(x0)
        f1_arr.append(g(x0))
        x0 = x1

        step = step + 1
        
        if step > N:
            flag=0
            break
            
        condition = abs(f(x1)) > e

    if flag==1:
        print('\nRequired root is: %0.8f' % x1)
    else:
        print('\nNot Convergent.')
    return c1_arr, f1_arr
```

```python

# Function to find the root 
def newtonRaphson( x ): 
    h = func(x) / derivFunc(x) 
    print('position', 'f(x)', 'dfx')
    c  = []
    fx = []
    dfx= []
    while abs(func(x)) >= 0.0001: #abs(h) >= 0.0001: 
        #h = func(x)/derivFunc(x) 
        
        # Multiple roots
        h = func(x)*derivFunc(x)/(derivFunc(x)**2 - func(x)*sderivFunc(x) )
        
        
        print('%.4f, %.4f, %.4f'%(x, func(x),derivFunc(x)) )
        c.append(x)
        fx.append(func(x))
        dfx.append(derivFunc(x))
        
        #x(i+1) = x(i) - f(x) / f'(x) 
        x = x - h 
      
    print("The value of the root is : ", 
                             "%.4f"% x) 
    return c, fx, dfx
```

```python
import numpy as np

def Newton_system(F, J, x, eps):
    """
    Solve nonlinear system F=0 by Newton's method.
    J is the Jacobian of F. Both F and J must be functions of x.
    At input, x holds the start value. The iteration continues
    until ||F|| < eps.
    """
    F_value = F(x)
    F_norm = np.linalg.norm(F_value, ord=2)  # l2 norm of vector
    iteration_counter = 0
    while abs(F_norm) > eps and iteration_counter < 100:
        delta = np.linalg.solve(J(x), -F_value)
        x = x + delta
        F_value = F(x)
        F_norm = np.linalg.norm(F_value, ord=2)
        iteration_counter += 1

    # Here, either a solution is found, or too many iterations
    if abs(F_norm) > eps:
        iteration_counter = -1
    return x, iteration_counter
```

```python

from numpy import cos, sin, pi, exp

def F(x):
    return np.array(
        [x[0]**2 - x[1] + x[0]*cos(pi*x[0]),
        x[0]*x[1] + exp(-x[1]) - 1.0/x[0]])

def J(x):
    return np.array(
        [[2*x[0] + cos(pi*x[0]) - pi*x[0]*sin(pi*x[0]), -1],
        [x[1] + 1./x[0]**(2), x[0] - exp(-x[1])]])
```

```python
expected = np.array([1, 0])
tol = 1e-4
x, n = Newton_system(F, J, x=np.array([2, -1]), eps=0.0001)
print (n, x)
error_norm = np.linalg.norm(expected - x, ord=2)
assert error_norm < tol, 'norm of error =%g' % error_norm
print ('norm of error =%g' % error_norm)
```
